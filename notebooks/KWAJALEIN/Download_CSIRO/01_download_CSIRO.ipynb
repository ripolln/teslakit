```python
# join files

rut = '/Users/albacid/Software/Bitbucket_repos/teslakit/notebooks/KWAJALEIN/Download_CSIRO/spec_points/spec_download/'


for yy in range(1979,2020):
    print(yy)
    
    for mm in range(1,13):
        print(mm)
        
        xds_m = xr.open_dataset(rut + '3034/ww3.{year}{month}_spec.nc'.format(year=str(yy), month=str(mm).zfill(2)))

        if 'efth' in xds_m.keys():               
            xds_m = xds_m.rename({'efth':'Efth','dpt':'depth','wnd':'u10m','wnddir':'udir','cur':'curr', 'curdir':'currdir'})

        if mm == 1:
            xds_out = xds_m
            
        else:
             xds_out = xr.concat([xds_out, xds_m], dim='time')

                
    print(xds_out)
    xds_out.to_netcdf(rut + str(yy) + '.nc')
    xds_out=None
```
